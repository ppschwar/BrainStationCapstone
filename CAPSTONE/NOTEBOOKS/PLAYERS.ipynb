Upon importing all 18 seasons, we combine all of the dataframes into a single composite in order to have all players, not repeated across various seasons.  Next we will read in our all star data. so we can then add this feature to the players dataframe above.

```python
all_stars = pd.read_csv("/Users/phil/Documents/Capstone_New/Data/NHL_AllStars_New.csv")
all_stars
```

```python
plt.figure()
plt.hist(all_stars['Count'])
plt.xlabel('Times a Player Made All Star Game')
plt.ylabel('Frequency')
plt.title('How often do NHL All-Stars Make the All Star Game?')
plt.show()
```

Above we can see a visual displaying that while a few players do make many all star games, almost half only play in the game 1 time, over the scope of seasons we are assessing.  This makes sense, when considering the quantity of the total player population relative to the number of all stars that play in any given season.
